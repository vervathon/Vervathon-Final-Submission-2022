```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()

X['StationId']=LE.fit_transform(X['StationId'])
X['StationName']=LE.fit_transform(X['StationName'])
X['Status']=LE.fit_transform(X['Status'])
X['State']=LE.fit_transform(X['State'])
X['Datetime']=LE.fit_transform(X['Datetime'])
X["AQI_bucket_calculated"]=LE.fit_transform(X["AQI_bucket_calculated"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=31,random_state=42)
```

```python
RF.fit(X_train,y_train)
```

```python
predR=RF.predict(X_test)
predR[:10]
```
